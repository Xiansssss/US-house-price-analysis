```python
import pandas as pd
from decimal import Decimal

# 读取第一个CSV文件
df1 = pd.read_csv("./raw-data/filtered-rental-data.csv")

# 读取第二个CSV文件
df2 = pd.read_csv("./raw-data/filtered_house_price_data.csv")

# 提取时间序列列的名称
time_series_columns1 = df1.columns[9:]  # 假设时间序列从第9列开始
time_series_columns2 = df2.columns[9:]

# 确保两个DataFrame包含相同的时间序列列
common_time_series_columns = list(set(time_series_columns1) & set(time_series_columns2))

# 选择所有需要的列，包括其他属性列
selected_columns = ['RegionID', 'SizeRank', 'RegionName', 'RegionType', 'StateName', 'State', 'City', 'Metro', 'CountyName'] + common_time_series_columns

# 确保选择的列在原始DataFrame中存在
selected_columns = [col for col in selected_columns if col in df1.columns and col in df2.columns]

# 更新DataFrame以包括所有需要的列
df1 = df1[selected_columns]
df2 = df2[selected_columns]

# 执行逐列的相除操作前，将df1中的值乘以12
df1[common_time_series_columns] = df1[common_time_series_columns] * 12

# 执行逐列的相除操作
result_df = df1.copy()
for col in common_time_series_columns:
    result_df[col] = df1[col].apply(Decimal) / df2[col].apply(Decimal)

# 按照时间序列的列名顺序进行排序
sorted_columns = selected_columns[:9] + sorted(common_time_series_columns)
result_df = result_df[sorted_columns]

# 将结果保存到新的CSV文件中
result_df.to_csv("new—result.csv", index=False)
```

```python
import pandas as pd

# 读取第一个CSV文件
df1 = pd.read_csv("./raw-data/filtered-rental-data.csv")

# 读取第二个CSV文件
df2 = pd.read_csv("./raw-data/filtered-house-price-data.csv")

# 创建一个新的DataFrame来存储筛选后的第二个文件的数据
filtered_df2 = pd.DataFrame(columns=df2.columns)

# 遍历第一个文件的每一行
for index, row in df1.iterrows():
    zipcode = row['RegionName']
    
    # 在第二个文件中使用布尔索引筛选出具有相同Zip Code的行
    matching_rows = df2[df2['RegionName'] == zipcode]
    
    # 将筛选后的行添加到新的DataFrame中
    filtered_df2 = filtered_df2.append(matching_rows, ignore_index=True)

# 将结果保存到新的CSV文件中
filtered_df2.to_csv("filtered_house_price_data.csv", index=False)
```
