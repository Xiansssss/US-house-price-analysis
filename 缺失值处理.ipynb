```python
import pandas as pd

# 指定CSV文件路径 
file_path = './data/result_data.csv'

# 使用na_values指定空值的表示
df = pd.read_csv(file_path, na_values=[""]) 
print(df.isnull().sum())
# 输出到同名CSV文件
df.to_csv(file_path, index=False)
```

```python
import pandas as pd

file_path = './data/result_data.csv'

# 读取CSV文件
df = pd.read_csv(file_path)

# 将空白值替换为单个空格
df.replace('', ' ', inplace=True)

# 将修改后的数据保存回原文件
df.to_csv(file_path, index=False)

print("替换完成！")
```

```python
import pandas as pd

file_path = './data/rental_data.csv'

# 读取CSV文件
df = pd.read_csv(file_path)

# 删除包含空值的行
df.dropna(axis=0, how='any', inplace=True)

# 将修改后的数据保存回原文件
df.to_csv(file_path, index=False)

print("删除完成！")
```

```python
import pandas as pd

file_path = './data/result_data.csv'

# 读取CSV文件
df = pd.read_csv(file_path)

# 删除每个属性值都是逗号的行
# 首先，将每个单元格的值替换为一个空字符串，然后判断整行是否全为空字符串，如果是则删除
df.replace(',', '', regex=True, inplace=True)
df.dropna(how='all', inplace=True)

# 将修改后的数据保存回原文件
df.to_csv(file_path, index=False)

print("删除逗号完成！")
```
